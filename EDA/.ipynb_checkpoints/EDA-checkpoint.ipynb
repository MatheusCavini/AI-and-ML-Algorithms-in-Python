## 4. Preparação dos dados

Primeiro, vamos selecionar apenas os atributos que de fato desejamos utilizar de nossa base de dados, eliminando todas as demais.

```python
#O comando .copy() no final garante que um novo dataframe seja gerado, e não apenas uma referência para o original
df = df[['coaster_name','Location', 'Status','Manufacturer', 'year_introduced', 'latitude', 'longitude', 
    'Type_Main', 'opening_date_clean','speed_mph', 'height_ft','Inversions_clean', 'Gforce_clean']].copy() 
```

```python
df.head(2)
```

Vamos também transformar a coluna de "Data de Abertura" para o formato de data padrão do Pandas. 

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

É conveniente também renomear as colunas para um único padrão.

```python
df = df.rename(columns={'coaster_name':'Coaster_Name',
                   'year_introduced':'Year_Introduced',
                   'opening_date_clean':'Opening_Date',
                   'speed_mph':'Speed_mph',
                   'height_ft':'Height_ft',
                   'Inversions_clean':'Inversions',
                   'Gforce_clean':'Gforce'})
```

Uma etapa importante é remover dados duplicados. Vamos remover aqui dados para os quais o nome, local e data de abertura da montanha russa são iguais.

```python
df = df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])].reset_index(drop=True).copy()
df.shape
```

## 5. Entendendo atributos individuais

Vamos primeiro tentar exibir em quais anos houve mais lançamentos de montanhas russas.

```python
df['Year_Introduced'].value_counts() #Conta quantos valores únicas para um atributo há no DF
```

Além de apenas visualizar os dados escritos, podemos começar a fazer plotagens simples com eles.

```python
ax = df['Year_Introduced'].value_counts().head(10).plot(kind='bar', title="Anos com maiores lançamentos de montanhas russas")
ax.set_xlabel("Ano de lançamento")
ax.set_ylabel("Nº de montanhas russas")
```

```python
ax1 = df['Speed_mph'].plot(kind='hist', bins=20)
ax1.set_title("Distribuição de velocidades")
ax1.set_xlabel("Velocidade (mph)")
ax1.set_ylabel("Nº de montanhas russas")
```

## 6. Explorando relações entre atributos

Vamos investigar por exemplo a relação entre a altura das montanhas russas e suas velocidades

```python
ax2 = df.plot(kind='scatter', x='Speed_mph', y='Height_ft')
ax2.set_title("Relação entre altura e velocidade")
ax2.set_xlabel("Velocidade (mph)")
ax2.set_ylabel("Altura (ft)")
```

A biblioteca Seaborn nos possibilita realizar algumas funções mais interessantes, como por exemplo utilizar o ano de lançamento como mapa de cor para a plotagem do gráfico, introduzindo visualmente uma nova informação.

```python
sns.scatterplot(x='Speed_mph', y='Height_ft', data=df, hue="Year_Introduced")
```

Outra função interessante é o pareamento entre diferentes dados em um único plot.

```python
sns.pairplot(df, vars=["Year_Introduced", "Speed_mph", "Height_ft", "Inversions", "Gforce"], hue="Type_Main")
```

Outro insight interessante pode ser analisar a correlação entre os pares de dados, isto como, como a variação de um influencia na variação do outro. Para isso, podemos utilizar a função .corr() do Pandas em conjunto com o mapa de calor do Seaborn.

```python
df_corr = df[["Year_Introduced", "Speed_mph", "Height_ft", "Inversions", "Gforce"]].dropna().corr()
```

```python
sns.heatmap(df_corr, annot=True)
```

## 7. Responder uma pergunta sobre os dados

Quais são os lugares com as montanhas russas mais velozes?

```python
ax3 = df.query("Location != 'Other'") \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean', 'count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh')

ax3.set_title("Média das velocidades de montanhas russas por local")
ax3.set_xlabel("Velocidade (mph)")
ax3.set_ylabel("Local")
```
