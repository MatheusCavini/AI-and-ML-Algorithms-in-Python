# Problemas de Busca em IA

No campo da inteliigncia artificial, os problemas de busca podem ser definidos como aqueles em que um agente  inserido num ambinete no qual pode tomar aes para atingir um objetivo. Consideramos a configurao do ambiente como estados, e conhecmos o modelo de transio de estados do problema, isto , para um dado estado atual e uma ao possvel, qual o prximo estado atingvel. 
Como base nisso, temos as seguintes definies:
- **Estados (espao de estados):** descrevem possveis situaes no ambiente. O espao de estados  o conjunto de todos os possveis estados.
- **Transies ou Aes:** possveis aes que levam de um estado ao outro. Apenas so consideradas transies diretas entre estados.
- **Custos:** valor relacionado a cada ao que identifica o quanto ela  custosa em algum sentido. Geralmente os algoritmos de busca tentam minimizar o custo.

Os problemas de busca possuem esse nome pois o objetivo do agente  buscar no mapeamento do espao de estados do problema qual a melhor sequncia de aes a ser tomada para atingir seu objetivo.

- **Um grafo de espao de estados** relaciona todos os possveis estados do ambiente atingveis a partir do estado inicial com uma funo de sucesso, que indica o prximo estado no caso de cada ao.
- Exemplo: grafo de estados de um rob aspirador onde as aes so go Left, go Right, Suck.



- **Caminho:** qualquer sequncia de estados conectados por uma sequncia de aes.
- **Soluo:** um caminho que leve do estado inicial ao estado objetivo.
- **Soluo tima:** uma soluo em que o custo total do caminho  o mnimo possvel.

Os algoritmos de soluo de problemas de busca so baseados em algoritmos de busca em rvore. Isso porque, no processo de expanso do grafo do espao de estados atravs das aes, pode-se criar uma rvore onde os ns so os estados, e dois ns sucessivos so ligados por uma ao que faz com o sistema passe do estado representado pelo n pai para o representado pelo n filho. Cada n na rvore de busca no representa porm apenas o estado atingido, mas sim todo o caminho (sequncia de aes) percorrido para chegar a ele.



Para estudar os algoritmos de busca, esse notebook usa o recurso de labirintos. Labirintos podem ser facilmente relacionados com grafos de espao de estados de problemas de busca, onde cada clula representa um estado: o estado de partida  a clula de "entrada" do labirinto, e o estado objetivo  a clula de "sada" do labirinto. Alm disso, para cada estado, so conhecidas as aes possveis de serem tomadas e seus estados resultante. Exemplo: se estou no estado representado pela clula $(1,2)$ e realizo a ao $\downarrow$, o estado resultante ser $(1,3)$.

## Algoritmos de Busca No Informada

Algoritmos de busca no informada so aqueles em que o agente realiza a busca sem dispor de qualquer informao adicional sobre o problema, apenas seu grafo de espao de estados.


### Depth-First Search (DFS)

O algoritmo de busca DFS  aquele onde a busca atravs da rvore resultante da expanso do grafo do espao de estados  feita de modo a primeiramente esgotar completamente um ramo da rvore at encontrar um n folha, para depois retroceder e explorar demais ramos.

![image.png](https://i.giphy.com/media/ii2FtzyirI4MXpoxX4/giphy.gif)

Veja que no algoritmo DFS precisamos definir uma direo preferencial de busca. No caso da rvore, se exploramos primeiro os ramos da esquerda ou da direita. Para o caso do labirinto, essa definio diz respeito a qual direo tentar seguir primeiro. Vamos adotar a ordem Oeste, Norte, Sul e Leste $(W,N,S,E)$

Para criar esse algoritmo, precisamos definir duas pilhas muito importantes:
- **Fronteira**: pilhando contendo as clulas que so listadas como "a exemplorar", por serem ns filhos de clular visitadas
- **Visitadas**: como o nome diz,  uma pilhas das clulas que j foram visitadas.

O algorimto consiste basicamente de, para cada estado atual, adicionar os prximos estados (clulas) possveis de serem explorados por aes viveis na pilha de fronteira, desde que esses estados no tenham sido visitados ainda. Ento, depois disso decidimos explorar o ltimo elemento da pilha. Repetimos isso at encontrar o objeto, ou um n folha, onde retornamos para o ltimo n com possibilidades de aes ainda no exploradas.

Com essas definies, vamos criar o algoritmo:

```python
def DFS(m, start, goal):
    startState = start 
    frontier = [startState]
    explored = []
    DFSBackPath = {}
    map = m.maze_map

    while len(frontier)>0:
        currentState = frontier.pop()
        explored.append(currentState)
        if currentState == goal:
            break
        for direction in 'ESNW':
            if map[currentState][direction] == True: #verifica se uma ao de movimentao na direo  vivel
                if direction == 'E':
                    childState = (currentState[0], currentState[1]+1)
                elif direction == 'S':
                    childState = (currentState[0]+1, currentState[1])
                elif direction == 'W':
                    childState = (currentState[0], currentState[1]-1)
                elif direction == 'N':
                    childState = (currentState[0]-1, currentState[1])
                
                if childState not in explored:
                    frontier.append(childState)
                    DFSBackPath[childState]=currentState

    DFSForwardPath = {}
    pathState = goal
    while pathState != startState:
        DFSForwardPath[DFSBackPath[pathState]]=pathState
        pathState = DFSBackPath[pathState]

    return DFSForwardPath
```

```python
m = maze(10,10)
m.CreateMaze(loopPercent=50)
path = DFS(m, (10,10), (1,1))
a = agent(m, footprints=True)
m.tracePath({a:path})
m.run()
```

Implementao alternativa que inclui visualizao da sequncia de busca do algoritmo:

```python
def VisualizerDFS(m, start, goal):
    startState = start 
    frontier = [startState]
    explored = []
    DFSBackPath = {}
    dSearch = []
    map = m.maze_map

    while len(frontier)>0:
        currentState = frontier.pop()
        dSearch.append(currentState)
        explored.append(currentState)

        if currentState == goal:
            break
        poss = 0
        for direction in 'ESNW':
            if map[currentState][direction] == True: #verifica se uma ao de movimentao na direo  vivel
                if direction == 'E':
                    childState = (currentState[0], currentState[1]+1)
                elif direction == 'S':
                    childState = (currentState[0]+1, currentState[1])
                elif direction == 'W':
                    childState = (currentState[0], currentState[1]-1)
                elif direction == 'N':
                    childState = (currentState[0]-1, currentState[1])
                
                if childState not in explored:
                    poss += 1
                    frontier.append(childState)
                    DFSBackPath[childState]=currentState
        if poss>1:
            m.markCells.append(currentState)

    DFSForwardPath = {}
    pathState = goal
    while pathState != startState:
        DFSForwardPath[DFSBackPath[pathState]]=pathState
        pathState = DFSBackPath[pathState]

    return dSearch, DFSBackPath, DFSForwardPath
```

```python
m = maze(25,25)
m.CreateMaze(loopPercent=25)
start = (25,25)
goal = (1,1)
dSearch, Bpath, Fpath = VisualizerDFS(m, start, goal)
a=agent(m,25,25,goal=goal,footprints=True,shape='square',color=COLOR.green)
m.tracePath({a:dSearch},showMarked=True,delay=10)

b=agent(m,1,1,goal=start,footprints=True,filled=True)
m.tracePath({b:Bpath}, delay=10)

c=agent(m,25,25,footprints=True,color=COLOR.yellow)
m.tracePath({c:Fpath}, delay=10)

m.run()
```

### Breadth-First Search (BFS)

O algoritmo de busca BFS  aquele onde a busca atravs da rvore resultante da expanso do grafo do espao de estados  feita de modo a visitar primeiramente todos os ns filhos do n de um determinado estado, para depois seguir para a "prxima gerao" de filhos. Isso signfica que um nvel da rvore de busca  visitado completamente antes de de partir para o prximo nvel.

![BFS GIF](https://i.giphy.com/media/iD6qdVmNOZH9frN6J4/giphy.gif)

Essa caracterstica nos fornece uma propriedade importante do algoritmo BFS: ele sempre retorna o menor caminho possvel at o objetivo, pois o caminho mais curto sempre est nos nveis superiores da rvore e eles so visitados primeiro.

A implementao do algoritmo  muito semelhante. Ele tambm conta as estruturas de dados que armazenam a fronteira e os ns visitados, e tambm precisa de uma direo preferencial de busca (aes a serem investigadas primeiro).

A principal diferena est no fato de que **enquanto a fronteira no DFS  implementada como uma Pilha, no BFS  implementada como uma fila.**

Isso signfica que os ns inseridos na lista de busca primeiro sero investigados primeiro, o que corresponde ao comportamento desejado de investigar um nvel completo primeiro.

Baseado nisso, temos a seguinte implementao para o BFS:

```python
def BFS(m, start, goal):
    startState = start 
    frontier = [startState]
    explored = []
    BFSBackPath = {}

    while len(frontier)>0:
        currentState = frontier.pop(0)
        explored.append(currentState)
        if currentState == goal:
            break
        for direction in 'ESNW':
            if m.maze_map[currentState][direction] == True: #verifica se uma ao de movimentao na direo  vivel
                if direction == 'E':
                    childState = (currentState[0], currentState[1]+1)
                elif direction == 'S':
                    childState = (currentState[0]+1, currentState[1])
                elif direction == 'W':
                    childState = (currentState[0], currentState[1]-1)
                elif direction == 'N':
                    childState = (currentState[0]-1, currentState[1])
                
                if childState not in explored:
                    frontier.append(childState)
                    BFSBackPath[childState]=currentState

    BFSForwardPath = {}
    pathState = goal
    while pathState != startState:
        BFSForwardPath[BFSBackPath[pathState]]=pathState
        pathState = BFSBackPath[pathState]

    return BFSForwardPath
```

```python
m = maze(10,10)
m.CreateMaze(loopPercent=50)
path = DFS(m, (10,10), (1,1))
a = agent(m, footprints=True)
m.tracePath({a:path})
m.run()
```

## Algoritmos de Busca Informada

Ao contrrio do mtodos de busca anteriores, onde o agente conhece apenas os estados inicial, final e o grafo de espao de estados do problema, as estratgias de busca informada tentam melhorar a eficincia dos algoritmos fornecendo dicas sobre a desejabilidade de diferentes estados.

Essa desejabilidade geralmente  ditada por um funo dependente do estado que chama **heurstica**, e pode ser entendida como um custo estimado para atingir o objetivo a partir daquele estado.

Assim, os algoritmos de busca informada em geral tentam expandir a busca com prioridade para aqueles estados de menor valor de heurstica, que em tese esto "mais prximos" em termos de custo do caminho restante at o objetivo.

### Algoritmo A*

O algoritmo de busca informada A* (A-star) busca expandir sua busca para um estado vizinho que minimize o valor da expresso:
$$ f(n) = g(n) + h(n) $$
Onde $g(n)$  o custo do caminho percorrido do estado inicial at aquele estado, e $h(n)$  a heurstica daquele estado, ou seja, o custo estimado dele at o estado final.

Para o tipo de problema que estamos trabalhando, de labirintos, as definies das funes so simples:
- $g(n)$: o nmero de passos dados at chegar na clula $n$
- $h(n)$:  distncia Manhattan  entre a clula $n$ e a clula final, isto , a soma das distncias horizontal e vertical (poderia ser a Euclidiana).

O resumo do algoritmo ento :
1. Em cada estado, atualizar a funo $f(n)$ para todos os estados vizinhos possveis (alcanveis por aes)
2. Se o valor de $f(n)$ atualizado for menor que um valor anteriormente calculado, considera-se esse como o novo valor.
3. Expande a busca para o estado com menor valor de $f(n)$

Para implementar, precisamos de uma estrutura de dados chamada **fila de prioridade**, onde os primeiros elementos a serem retirados da fila so aqueles com maior (ou menor) valor de prioridade.


```python
from queue import PriorityQueue
```

```python
#Definio da funo de Heurstica com a Distncia Manhattan
def h(state, goal):
    x1, y1 = state
    x2, y2 = goal
    return abs(x1-x2) + abs(y1-y2)

#Algoritmo A*
def AStar(m, start, goal):
    startState = start 
    g_scores = {state:float('inf') for state in m.grid}
    g_scores[start] = 0
    f_scores = {state:float('inf') for state in m.grid}
    f_scores[start] = h(start, goal)

    to_visit = PriorityQueue()
    to_visit.put((h(start, goal), h(start, goal), start))
    AStarBackPath = {}

    while not to_visit.empty():
        currentState = to_visit.get()[2]
        if currentState == goal:
            break
        
        for direction in 'ESNW':
            if m.maze_map[currentState][direction] == True: #verifica se uma ao de movimentao na direo  vivel
                if direction == 'E':
                    childState = (currentState[0], currentState[1]+1)
                elif direction == 'S':
                    childState = (currentState[0]+1, currentState[1])
                elif direction == 'W':
                    childState = (currentState[0], currentState[1]-1)
                elif direction == 'N':
                    childState = (currentState[0]-1, currentState[1])

                temp_g_score = g_scores[currentState]+1
                temp_f_score = temp_g_score + h(childState, goal)

                if temp_f_score < f_scores[childState]:
                    g_scores[childState] = temp_g_score
                    f_scores[childState] = temp_f_score
                    to_visit.put((temp_f_score, h(childState, goal), childState))
                    AStarBackPath[childState]=currentState

    

    AStarForwardPath = {}
    pathState = goal
    while pathState != startState:
        AStarForwardPath[AStarBackPath[pathState]]=pathState
        pathState = AStarBackPath[pathState]

    return AStarForwardPath
```

```python
m = maze(10,10)
m.CreateMaze(loopPercent=50)
path = AStar(m, (10,10), (1,1))
a = agent(m, footprints=True)
m.tracePath({a:path})
m.run()
```

Implementao alternativa que inclui a visualizao da sequncia de busca do algoritmo:
