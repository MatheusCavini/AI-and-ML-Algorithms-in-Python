# Implementação do Zero

```python
import numpy as np
from collections import Counter
```

```python
def euclidian_distance(x1, x2):
    dist = np.sqrt(np.sum((x1-x2)**2))
    return dist
```

```python
class KNN:
    def __init__(self, k):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y
        
    def predict(self, X):
        predictions = [self._predict_single_sample(x) for x in X]
        return predictions
        
    def _predict_single_sample(self, x):
        distances = [euclidian_distance(x, x_train) for x_train  in self.X_train]
        k_idxs = np.argsort(distances)[:self.k]
        k_labels = [self.y_train[i] for i in k_idxs]
        output_label = Counter(k_labels).most_common()
        return output_label[0][0]
```

# Teste

```python
#!pip install scikit-learn
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
cmap = ListedColormap(["#FF0000", "#00FF00", "#0000FF"])
```

```python
iris = datasets.load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size = 0.2)
```

```python
plt.figure()
plt.scatter(X[:,2], X[:,3], c=y, cmap=cmap, edgecolor='k', s=20)
plt.show()
```

```python
clf =  KNN(k=5)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
print(predictions)
```

```python
acc = np.sum(predictions == y_test)/len(y_test)
print(acc)
```

```python
plt.scatter(X_test[:,2], X_test[:, 3], c=predictions, cmap=cmap, edgecolor='k', s=20)
plt.show()
```
